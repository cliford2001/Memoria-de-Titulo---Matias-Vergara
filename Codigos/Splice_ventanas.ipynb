```python
import pandas as pd

# Función para procesar eventos SE
def SE_event(event):
    chr = event['chr']
    name = event['geneSymbol']
    score = event['IncLevelDifference']
    strand = event['strand']

    # Obtener las coordenadas para analizar en el exón upstream
    up_start = event['upstreamEE'] - 50
    up_end = event['upstreamEE'] + 250

    # Obtener las coordenadas para analizar en el exón saltado
    skip_start1 = event['exonStart_0base'] - 250
    skip_end1 = event['exonStart_0base'] + 50
    skip_start2 = event['exonEnd'] - 50
    skip_end2 = event['exonEnd'] + 250

    # Obtener las coordenadas para analizar en el exón downstream
    down_start = event['downstreamES'] - 250
    down_end = event['downstreamES'] + 50

    # Generar una lista con las 4 regiones
    all_regions = [
        [chr, up_start, up_end, 'upstream', score, strand],
        [chr, skip_start1, skip_end1, 'skip1', score, strand],
        [chr, skip_start2, skip_end2, 'skip2', score, strand],
        [chr, down_start, down_end, 'downstream', score, strand]
    ]

    return all_regions

# Leer el archivo de entrada
file_path = "/content/drive/MyDrive/U/11° Semestro y ultimo espero/Tesis/MDA/codigo/datos_filtrados_se.tsv"
data = pd.read_csv(file_path, sep="\t")

# Crear una lista para almacenar los resultados
all_results = []

# Iterar sobre cada fila del archivo y aplicar la función SE_event
for index, row in data.iterrows():
    results = SE_event(row)
    all_results.extend(results)

# Convertir los resultados a un DataFrame
columns = ['chr', 'start', 'end', 'region', 'score', 'strand']
results_df = pd.DataFrame(all_results, columns=columns)

# Guardar los resultados en un archivo de salida
output_file_path = "/content/drive/MyDrive/U/11° Semestro y ultimo espero/Tesis/MDA/codigo/output_SE_events.bed"
results_df.to_csv(output_file_path, sep="\t", index=False, header=False)

print(f"Resultados guardados en {output_file_path}")
```

```python
import pandas as pd

# Función para procesar eventos MXE
def MXE_event(event):
    chr = event['chr']
    name = event['geneSymbol']
    score = event['IncLevelDifference']
    strand = event['strand']

    # Obtener las coordenadas para analizar en el exón upstream
    up_start = event['upstreamEE'] - 50
    up_end = event['upstreamEE'] + 250

    # Obtener las coordenadas para analizar en el primer exón
    first_start1 = event['1stExonStart_0base'] - 250
    first_end1 = event['1stExonStart_0base'] + 50
    first_start2 = event['1stExonEnd'] - 50
    first_end2 = event['1stExonEnd'] + 250

    # Obtener las coordenadas para analizar en el segundo exón
    second_start1 = event['2ndExonStart_0base'] - 250
    second_end1 = event['2ndExonStart_0base'] + 50
    second_start2 = event['2ndExonEnd'] - 50
    second_end2 = event['2ndExonEnd'] + 250

    # Obtener las coordenadas para analizar en el exón downstream
    down_start = event['downstreamES'] - 250
    down_end = event['downstreamES'] + 50

    # Generar una lista con todas las regiones
    all_regions = [
        [chr, up_start, up_end, 'upstream', score, strand],
        [chr, first_start1, first_end1, 'first1', score, strand],
        [chr, first_start2, first_end2, 'first2', score, strand],
        [chr, second_start1, second_end1, 'second1', score, strand],
        [chr, second_start2, second_end2, 'second2', score, strand],
        [chr, down_start, down_end, 'downstream', score, strand]
    ]

    return all_regions

# Leer el archivo de entrada
file_path = "/content/drive/MyDrive/U/11° Semestro y ultimo espero/Tesis/MDA/codigo/datos_filtrados_mxe.tsv"
data = pd.read_csv(file_path, sep="\t")

# Crear una lista para almacenar los resultados
all_results = []

# Iterar sobre cada fila del archivo y aplicar la función MXE_event
for index, row in data.iterrows():
    results = MXE_event(row)
    all_results.extend(results)

# Convertir los resultados a un DataFrame
columns = ['chr', 'start', 'end', 'region', 'score', 'strand']
results_df = pd.DataFrame(all_results, columns=columns)

# Guardar los resultados en un archivo de salida
output_file_path = "/content/drive/MyDrive/U/11° Semestro y ultimo espero/Tesis/MDA/codigo/output_MXE_events.bed"
results_df.to_csv(output_file_path, sep="\t", index=False, header=False)

print(f"Resultados guardados en {output_file_path}")
```

```python
import pandas as pd

# Función para procesar eventos RI
def RI_event(event):
    chr = event['chr']
    name = event['geneSymbol']
    score = event['IncLevelDifference']
    strand = event['strand']

    # Obtener las coordenadas para analizar en el exón upstream
    up_start = event['upstreamES']
    up_end = event['upstreamES'] + 50

    # Obtener las coordenadas para analizar en el intrón retenido
    ri_start = event['upstreamEE'] - 50
    ri_end = event['downstreamES'] + 50

    # Obtener las coordenadas para analizar en el exón downstream
    down_start = event['downstreamEE'] - 50
    down_end = event['downstreamEE']

    # Generar una lista con todas las regiones
    all_regions = [
        [chr, up_start, up_end, 'upstream', score, strand],
        [chr, ri_start, ri_end, 'ri', score, strand],
        [chr, down_start, down_end, 'downstream', score, strand]
    ]

    return all_regions

# Leer el archivo de entrada
file_path = "/content/drive/MyDrive/U/11° Semestro y ultimo espero/Tesis/MDA/codigo/datos_filtrados_ri.tsv"
data = pd.read_csv(file_path, sep="\t")

# Crear una lista para almacenar los resultados
all_results = []

# Iterar sobre cada fila del archivo y aplicar la función RI_event
for index, row in data.iterrows():
    results = RI_event(row)
    all_results.extend(results)

# Convertir los resultados a un DataFrame
columns = ['chr', 'start', 'end', 'region', 'score', 'strand']
results_df = pd.DataFrame(all_results, columns=columns)

# Guardar los resultados en un archivo de salida
output_file_path = "/content/drive/MyDrive/U/11° Semestro y ultimo espero/Tesis/MDA/codigo/output_RI_events.bed"
results_df.to_csv(output_file_path, sep="\t", index=False, header=False)

print(f"Resultados guardados en {output_file_path}")
```

```python
import pandas as pd
import sys

# Función para procesar eventos ASS (A3SS y A5SS)
def ASS_event(event, TypeEvent):
    chr = event['chr']
    name = event['geneSymbol']
    score = event['IncLevelDifference']
    strand = event['strand']

    if (((strand == '+') & (TypeEvent == 'A3SS')) | ((strand == '-') & (TypeEvent == 'A5SS'))):  # izquierda a derecha
        # Coordenadas para el exón flanqueante
        flank_start1 = event['flankingES']
        flank_end1 = event['flankingES'] + 50
        flank_start2 = event['flankingEE'] - 50
        flank_end2 = event['flankingEE'] + 250

        # Coordenadas para el sitio de splicing
        ss_start1 = event['longExonStart_0base'] - 250
        ss_end1 = event['longExonStart_0base'] + 50
        ss_start2 = event['shortES'] - 50
        ss_end2 = event['shortES'] + 50
        ss_start3 = event['longExonEnd'] - 50
        ss_end3 = event['longExonEnd']

    elif (((strand == '-') & (TypeEvent == 'A3SS')) | ((strand == '+') & (TypeEvent == 'A5SS'))):  # derecha a izquierda
        # Coordenadas para el exón flanqueante
        flank_start1 = event['flankingEE'] - 50
        flank_end1 = event['flankingEE']
        flank_start2 = event['flankingES'] - 250
        flank_end2 = event['flankingES'] + 50

        # Coordenadas para el sitio de splicing
        ss_start1 = event['longExonEnd'] - 50
        ss_end1 = event['longExonEnd'] + 250
        ss_start2 = event['shortEE'] - 50
        ss_end2 = event['shortEE'] + 50
        ss_start3 = event['longExonStart_0base']
        ss_end3 = event['longExonStart_0base'] + 50

    else:
        sys.exit("Error in Strand information: strand = {}".format(strand))

    # Generar una lista con todas las regiones
    all_regions = [
        [chr, flank_start1, flank_end1, 'flank1', score, strand],
        [chr, flank_start2, flank_end2, 'flank2', score, strand],
        [chr, ss_start1, ss_end1, 'ss1', score, strand],
        [chr, ss_start2, ss_end2, 'ss2', score, strand],
        [chr, ss_start3, ss_end3, 'ss3', score, strand]
    ]

    return all_regions

# Leer el archivo de entrada
file_path = "/content/drive/MyDrive/U/11° Semestro y ultimo espero/Tesis/MDA/codigo/datos_filtrados_ASS.tsv"
data = pd.read_csv(file_path, sep="\t")

# Crear una lista para almacenar los resultados
all_results = []

# Iterar sobre cada fila del archivo y aplicar la función ASS_event
for index, row in data.iterrows():
    results = ASS_event(row, row['Tipo'])
    all_results.extend(results)

# Convertir los resultados a un DataFrame
columns = ['chr', 'start', 'end', 'region', 'score', 'strand']
results_df = pd.DataFrame(all_results, columns=columns)

# Guardar los resultados en un archivo de salida
output_file_path = "/content/drive/MyDrive/U/11° Semestro y ultimo espero/Tesis/MDA/codigo/output_ASS_events.bed"
results_df.to_csv(output_file_path, sep="\t", index=False, header=False)

print(f"Resultados guardados en {output_file_path}")
```
